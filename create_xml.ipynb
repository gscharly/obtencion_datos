```python
csv_file = open("stops_web.csv","r")
csv_reader = csv.reader(csv_file, delimiter=',', lineterminator='\n')
```

```python
g = Graph()

#sch = Namespace('https://schema.org')
gen = Namespace('http://www.transporte_madrid.es#')
est = Namespace('http://www.transporte_madrid.es/parada#')
lin = Namespace('http://www.transporte_madrid.es/linea#')
tra = Namespace('http://www.transporte_madrid.es/transporte#')
dbo = Namespace('http://http://dbpedia.org/ontology/Place#')
geo = Namespace("http://www.w3.org/2003/01/geo/wgs84_pos#")
```

Básico:

```python
#Pasamos la cabecera
next(csv_reader)
for row in csv_reader:
    transport = URIRef(tra + row[0])
    line = URIRef(lin + row[0] + '_' + row[1])
    g.add ( (line, gen.ofTransport, transport))
    stop = URIRef(est + row[3])
    g.add ( (stop, gen.ofLine, line) )
    g.add ( (stop, FOAF.name, Literal(row[5])) )
    g.add((stop, gen.isNumber, Literal(str(int(row[2].split('_')[1])+1))))
    g.add((stop, gen.stopCode, Literal(4)))
    g.add((stop, dbo.address, Literal(row[6])))
    g.add((stop, geo.lat, Literal(row[7])))
    g.add((stop, geo.long, Literal(row[8])))
    g.add((stop, gen.zone, Literal(row[9])))
    g.add((stop, gen.url, Literal(row[10])))
    g.add((stop, gen.locationType, Literal(row[11])))
    g.add((stop, gen.parentStation, Literal(row[12])))
    g.add((stop, OWL.TimeZone, Literal(row[13])))
    g.add((stop, gen.wheelchairBoarding, Literal(row[14])))
    
    
g.serialize(destination ="prueba_estaciones_basico.xml", format = "xml")
```

```python
#Pasamos la cabecera
next(csv_reader)
for row in csv_reader:
    
    transport = URIRef(tra + row[0])
    line = URIRef(lin + row[0] + '_' + row[1])

    nodes_line = [l for l in g.objects(line, gen.hasStations)]
    if len(nodes_line) == 0:
        seq = BNode()
        g.add((line, gen.hasStations, seq))
        g.add((seq, RDF.type, RDF.Seq))
    else:
        seq = nodes_line[0]
    
    #g.add ( (line, gen.ofTransport, transport))
    stop = URIRef(est + row[3])
    #g.add((stop, gen.ofLine, seq))
    #g.add((seq, URIRef(gen + row[2]), stop))
    g.add( (seq, URIRef('http://www.w3.org/1999/02/22-rdf-syntax-ns#_' + str(int(row[2].split('_')[1])+1)), stop)) 
    #g.add ( (stop, gen.ofLine, line) )
    g.add ( (stop, FOAF.name, Literal(row[5])) )
    # AÑADIR RESTO DE ATRIBUTOS
    
g.serialize(destination ="prueba_estaciones.xml", format = "xml")
```

```python
for person in g.predicates(gen.hasStations):
    print(person)
```

```python
for s,p,o in g.triples((None, gen.hasStations, None)):
    print(s,p,o)
```

```python
import rdflib

book            = rdflib.URIRef(u'urn:isbn:978-3-16-148410-0')
bookType        = rdflib.URIRef(u'urn:publishing:schema:book')
authorPredicate = rdflib.URIRef(u'http://purl.org/dc/terms/creator')
joyce           = rdflib.URIRef(u'urn:foaf:registered-people:joyce')
kelly           = rdflib.URIRef(u'http://twitter.org/nutjob4life')
mattmann        = rdflib.URIRef(u'http://people.apache.org/mattmann')

# We have a book
g = rdflib.Graph()
g.add((book, rdflib.RDF.type, bookType))

# It has a collection of authors
bag = rdflib.BNode()
g.add((book, authorPredicate, bag))

# The collection is a "Bag"
g.add((bag, rdflib.RDF.type, rdflib.RDF.Bag))

# The bag has 3 authors
g.add((bag, rdflib.URIRef(u'http://www.w3.org/1999/02/22-rdf-syntax-ns#_1'), joyce))
g.add((bag, rdflib.URIRef(u'http://www.w3.org/1999/02/22-rdf-syntax-ns#_2'), kelly))
g.add((bag, rdflib.URIRef(u'http://www.w3.org/1999/02/22-rdf-syntax-ns#_3'), mattmann))

# That's it.
print(g.serialize())
```
