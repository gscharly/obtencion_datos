## Unificación de fuentes

```python
def clean_station(station):
    # Tildes
    clean_station = unidecode.unidecode(station)
    # Pasar a mayus
    clean_station = clean_station.upper()
    # Estaciones de ML con (ida), (vuelta)
    clean_station = clean_station.split('(')[0]
    # Espacios innecesarios
    clean_station = clean_station.strip()
    # Otros procesados
    clean_station = clean_station.replace('RDA.', 'RONDA')
    clean_station = clean_station.replace('AVDA.', 'AVENIDA')
    clean_station = clean_station.replace("'", '')
    clean_station = clean_station.replace(" - ", '-')
    
    if clean_station == 'MOSTOLES-EL SOTO':
        clean_station = 'MOSTOLES EL SOTO'
    elif clean_station == 'BARRIAL-C.COM.POZUELO EL':
        clean_station = 'EL BARRIAL-C.C. POZUELO'
    elif clean_station == 'PUERTO NAVACERRADA':
        clean_station = 'PUERTO DE NAVACERRADA'
    elif clean_station == 'ATOCHA RENFE':
        clean_station = 'ATOCHA-RENFE'
    elif clean_station == 'AEROPUERTO T1 T2 T3':
        clean_station = 'AEROPUERTO T1-T2-T3'
    return clean_station

def clean_web_stations(stations_web):
    stations_web_cp = stations_web.copy()
    for t, t_lines in stations_web_cp.items():
        for line, stations in t_lines.items():
            for i in range(len(stations)):
                stations_web_cp[t][line][i] = clean_station(stations[i])
    return stations_web_cp
```

__Web__

Transformamos el diccionario obtenido de la web en una lista de tuplas (transporte, estación, línea, orden)

```python
def from_dict_to_list(stations_web):
    stations_web_list = []
    for t in crtm.TRANSPORT_TYPES:
        for line,stations in stations_web[t].items():
            for s in stations:
                stations_web_list.append((t,s,line,line + '_' + str(stations.index(s))))
    return stations_web_list
```

```python
stations_web_list = from_dict_to_list(stations_web)
```

Con esta lista se construye un dataframe que será la base de partida. Habrá tantos registros como estaciones dentro de las líneas.

También se limpian los datos en este punto: tildes, todo en mayus, cambios en algunos nombres...

```python
def clean_dataframe(pd_df, web=True):
    if web:
        pd_df['transportmean_name'] = pd_df['transportmean_name'].apply(lambda x: 'METRO' if x=='metro' else('ML' if x=='metro-ligero' else 'CR'))
        pd_df['line_number'] = pd_df['line_number'].apply(lambda x: x.replace('-', ''))
        pd_df['order_number'] = pd_df['order_number'].apply(lambda x: x.replace('-', ''))
    pd_df['stop_name'] = pd_df['stop_name'].apply(clean_station)
    return pd_df

def web_dataframe(stations_web):
    # From dict to list
    stations_web_list = from_dict_to_list(stations_web)
    # Create Pandas df
    pd_stations_web = pd.DataFrame(stations_web_list, columns=['transportmean_name','stop_name', 'line_number', 'order_number'])
    # Cleaning
    pd_stations_web = clean_dataframe(pd_stations_web)
    return pd_stations_web
```

__Stops__

Hay que arreglar algunas paradas de cercanías: CASAR EL -> EL CASAR

location_type:

- 0: paradas
- 1: estaciones. Pueden incluir paradas o accesos
- 2: entradas/salidas

```python
def groupby_stations(pd_stops):
    pd_stops_group = pd_stops.groupby('stop_name').first()
    return pd_stops_group

def stops_transport_dataframe(transport):
    pd_stops = pd.read_csv("stops-{}.txt".format(transport))
    # Cleaning
    pd_stops = clean_dataframe(pd_stops, web=False)
    # Only stops
    pd_stops_par = pd_stops[pd_stops.location_type==0].copy()
    # Stations and accesses
    #pd_stops_est_acc = pd_stops[(pd_stops.location_type==1) | (pd_stops.location_type==2)].copy()
    # Process cercanias
    if transport == "cercanias-renfe":
        reverse_articles = lambda x: x.split()[-1] + ' ' + ' '.join(x.split()[:-1]) if x.split()[-1] in ['EL', 'LA', 'LAS', 'LOS'] else x
        pd_stops_par['stop_name'] = pd_stops_par['stop_name'].apply(reverse_articles)
    pd_stops_par = groupby_stations(pd_stops_par)
    
    return pd_stops_par
```
